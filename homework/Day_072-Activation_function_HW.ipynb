# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
# ReLU

x = np.arange(-10, 10, 0.01)
y = list(map(lambda x:x if x > 0 else 0,x))

# 繪圖
plt.plot(x,y)
plt.grid(color = 'b', linewidth='0.1', linestyle='--')
```

```python
# dReLU 一階導數
dev_y = list(map(lambda x:1 if x > 0 else 0, x))

# 繪圖
plt.plot(x,dev_y)
plt.grid(color = 'b', linewidth='0.1', linestyle='--')
```
